To address my specific question, I will use a K-Nearest Neighbours (K-NN) classification model. K-NN is an appropriate choice because my response variable, **subscribe**, is binary (TRUE/FALSE), and K-NN naturally handles categorical classification outcomes. My explanatory variables (**played_hours**, **Age**, and **experience**) consist of numeric and categorical predictors that can be easily incorporated into a K-NN framework.

#### Why is K-NN the right choice?
* K-NN is specifically designed to classify observations into categories such as subscribers vs. non-subscribers
* K-NN does not assume linear relationships between the predictors and the response, so it is more flexible when modelling human behaviour like playing habits
* It is intuitive, K-NN predicts a player's subscripton status based on similar players

#### Assumptions
* For K-NN to work, the variables must be scaled appropriately because it assumes that similarity in predictors reflect similarity in the outcome
* K-NN also depends heavily on the distribution of the data, so the training data must have enough observations for all experience levels and subscription status

#### Potential limitations
* Variables measured on diffeent scales can easily distort distances, so needs to be standardised
* Results can vary depending on the value of k chosen

#### Comparing & selecting the model
I will evaluate the model using cross-validation to compare the classification accuracy (and maybe the precision or recall) across different values of k

Some specific measures may include:
* Using a 5-fold or 10-fold cross validation to get stable performance estimates
* tuning the k parameter over a grid (i.e., k = 1 to 25
* the value of k that produces the highest average accuracy across the folds will be used for the final model

#### Processing the data to apply model
1. Scaling


Because K-NN relies on distances, I will scale numeric predictors (Age and played_hours) based only on the training data, then apply the same scaling parameters to the test data.

2. Splitting the data
* I will split the data after all minimal wrangling/scaling using only the training data. <br>

Planned split will look like the following to provide a good balance:

Training set: 80% <br>
Testing set: 20%

3. Cross-Validation
With the training set only:
* perform a 5-fold cross validation
* tune k
* choose the best model
* evaluate the final performance on untouched test set

A  validation set is not needed because cross-validation in the training set will take care of the model selection

4. Final model evaluation
   
Once the best model is chosen by cross-validation, I will:
* Fit the model to the full training set
* Evaluate the accuracy on the separate 20% test set
