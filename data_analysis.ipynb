```r
# Wrangling the data to make it work
library(tidyverse)
library(dplyr)

players <- read_csv("data/players.csv")

tidy_players <- players |>
    mutate(
        experience = as.factor(experience),
        subscribe = as.factor(subscribe)
    )
tidy_players
```

```r
# Computing the mean for each quantitative variable

mean_hours <- tidy_players |>
    pull(played_hours) |>
    mean(na.rm = TRUE)

mean_age <- tidy_players |>
    pull(Age) |>
    mean(na.rm = TRUE)

# Reporting values in table format
mean_table <- tibble(
    mean_played_hours = mean_hours,
    mean_player_age = mean_age)
mean_table
```

```r
library(tidyverse)
library(dplyr)

# Making exploratory visualisations
## Plot 1: Distribution of Played Hours
options(repr.plot.width = 10, repr.plot.height = 10)
plot_1 <- ggplot(tidy_players, aes(x = played_hours)) +
    geom_histogram(bins = 30) +
    labs(title = "Distribution of Hours Played", x = "Hours Played", y = "Count") +
    theme(text = element_text(size = 20))
plot_1
```
